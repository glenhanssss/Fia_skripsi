```python
# load data Kapur sebelum

import pandas as pd
import glob

# folder menyimpan file excel
path = r'..\path_sebelum'

# Membuat pola pencarian untuk file Data_Sampel1.xlsx hingga Data_Sampel9.xlsx
files_pattern = path + "/Data_Sampel[10-18].xlsx"

# Mencari file sesuai dengan pola
all_files = glob.glob(files_pattern)

li = []
for filename in all_files:
    df = pd.read_excel(filename, index_col=None, header=0)
    li.append(df)

# gabungin semua file excel
df = pd.concat(li, axis=0, ignore_index=True)

# hapus column Intensitas Warna biar urutan nya ga jelek kaya kamu
df.drop('Intensitas Warna', axis=1, inplace=True)

# Mengubah nama semua kolom df_Kapur_sebelum
df = df.rename(columns=lambda x: x + '_Kapur_sebelum')

# tambah column Intensitas Warna biar urutan nya dari 1 sampe 2304
df.insert(0, 'Intensitas_Warna', range(1, len(df) + 1))

# rename nama df_Kapur_sebelum
df_Kapur_sebelum = df

# display df_Kapur_SESUDAH
df_Kapur_sebelum
```

```python
# load data Kapur SESUDAH

import pandas as pd
import glob

# folder menyimpan file excel
path = r'..\path_sesudah'

# Membuat pola pencarian untuk file Data_Sampel1.xlsx hingga Data_Sampel9.xlsx
files_pattern = path + "/Data_Sampel[10-18].xlsx"

# Mencari file sesuai dengan pola
all_files = glob.glob(files_pattern)

li = []
for filename in all_files:
    df = pd.read_excel(filename, index_col=None, header=0)
    li.append(df)

# gabungin semua file excel
df = pd.concat(li, axis=0, ignore_index=True)

# hapus column Intensitas Warna biar urutan nya ga jelek kaya kamu
df.drop('Intensitas Warna', axis=1, inplace=True)

# Mengubah nama semua kolom df_Kapur_SESUDAH
df = df.rename(columns=lambda x: x + '_Kapur_SESUDAH')

# tambah column Intensitas Warna biar urutan nya dari 1 sampe 2304
df.insert(0, 'Intensitas_Warna', range(1, len(df) + 1))

# rename nama df_Kapur_SESUDAH
df_Kapur_SESUDAH = df

# display df_Kapur_SESUDAH
df_Kapur_SESUDAH
```

```python
# merge data sebelum dan SESUDAH

# merge df_Kapur_sebelum dan df_Kapur_SESUDAH
df = pd.merge(df_Kapur_sebelum, df_Kapur_SESUDAH, on="Intensitas_Warna")
df
```

```python
# mencari rata-rata untuk membuat visualisasi perbandingan

# hapus column Intensitas Warna karna udah ga guna
df_average = df
df_average.drop('Intensitas_Warna', axis=1, inplace=True)

# Menghitung rata-rata dari setiap kolom
column_averages = df_average.mean()

# Membuat DataFrame baru dengan nilai rata-rata kolom sebelumnya
df_average = pd.DataFrame(column_averages).transpose()

# membulatkan hasil rata-rata menjadi 4 angka dibelakang koma
df_average = df_average.round(4)

# display dataframe
df_average
```

```python
# split data into 
data_R = {
    'AVG_Kapur_sebelum': df_average["Jumlah titik-R_Kapur_sebelum"],
    'AVG_Kapur_SESUDAH': df_average["Jumlah titik-R_Kapur_SESUDAH"]
}
df_R = pd.DataFrame(data_R)

data_G = {
    'AVG_Kapur_sebelum': df_average["Jumlah titik-G_Kapur_sebelum"],
    'AVG_Kapur_SESUDAH': df_average["Jumlah titik-G_Kapur_SESUDAH"]
}
df_G = pd.DataFrame(data_G)

data_B = {
    'AVG_Kapur_sebelum': df_average["Jumlah titik-B_Kapur_sebelum"],
    'AVG_Kapur_SESUDAH': df_average["Jumlah titik-B_Kapur_SESUDAH"]
}
df_B = pd.DataFrame(data_B)

data_gray = {
    'AVG_Kapur_sebelum': df_average["Jumlah titik-gray_Kapur_sebelum"],
    'AVG_Kapur_SESUDAH': df_average["Jumlah titik-gray_Kapur_SESUDAH"]
}
df_gray = pd.DataFrame(data_gray)

# gabung lagi dataframe nya
df = pd.concat([df_R, df_G, df_B, df_gray], ignore_index=True)

# display data
df
```

```python
# adding column warna
df.insert(0, "warna", ["R", "G", "B", "gray"], True)

# display data
df
```

```python
# visualize data

import matplotlib.pyplot as plt

# Grouped bar plot
ax = df.plot(x='warna', kind='bar')

# Add mean values as text on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10),
                textcoords='offset points', rotation='vertical')  # Rotate text vertically

# Add labels and title
plt.xlabel('warna')
plt.ylabel('Mean Value')
plt.title('Kapur Visualization')

# Show plot
plt.show()
```
